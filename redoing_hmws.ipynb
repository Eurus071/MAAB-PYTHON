```python
def uncommon(list1, list2):
    uncommon=[]

    for x in list1:
        if x not in list2:
            uncommon.append(x)
    
    for x in list2:
        if x not in list1:
            uncommon.append(x)
    
    return uncommon

list1=[1, 2, 3, 4]
list2=[3, 4, 5, 6]

uncommon(list1, list2)
```

```python
"""
### Bonus Challenge
Task: Create a simple text-based Rock, Paper, Scissors game where the player plays against the computer.
- The computer randomly chooses `rock`, `paper`, or `scissors` using `random.choice()`.
- The player enters their choice.
- Display the winner and keep track of scores for the player and the computer.
- First to 5 points wins the match.

"""
import random
choices=['rock', 'paper', 'scissors']

human_score=0
comp_score=0


while human_score<5 and comp_score<5:
    human_choice = input("Enter your choice (rock, paper, scissors): ").lower()
    comp_choice = random.choice(choices)

    print(f"Computer chose: {comp_choice}")

    if human_choice==comp_choice:
         print("No one won")
    elif human_choice=='paper' and comp_choice=='rock':
         human_score+=1
    elif human_choice=='rock' and comp_choice=='scissors':
         human_score+=1
    elif human_choice=='scissors' and comp_choice=='paper':
        human_score+=1
    else:
        comp_score+=1
    
    print(f"Score -> You: {human_score} | Computer: {comp_score}")
    print("-" * 30)

print("Game over")





```

```python
"""
**3.** `txt` nomli string saqlovchi o'zgaruvchi berilgan. `txt`dagi har uchinchi belgidan
 keyin pastgi chiziqcha (underscore) qo'yilsin. Agar belgi unli harf yoki orqasidan ostki
 chiziqcha qo'yilgan harf bo'lsa, ostki chiziqcha keyingi harfdan keyin qo'yilsin.
 Agar belgi satrdagi oxirgi belgi bo'lsa chiziqcha qo'yilmasin.


input: hello
output: hel_lo

input: assalom
output: ass_alom

input: abcabcdabcdeabcdefabcdefg
output: abc_abcd_abcdeab_cdef_abcdefg

"""

def add_underscore(txt):
    vowels='aeiou'
    count=0
    result=''

    for i in range(len(txt)):
        result+=txt[i]
        count+=1

        if count==3 and i != (len(txt)-1):
            if txt[i] not in vowels:
                result+='_'
                count=0
    return result

txt=input("Enter your string: ")
add_underscore(txt)

```

```python
"""
**4. Number Guessing Game**
Create a simple number guessing game.
- The computer randomly selects a number between 1 and 100. 
- If the guess is high, print "Too high!". 
- If the guess is low, print "Too low!". 
- If they guess correctly, print "You guessed it right!" and exit the loop.
- The player has 10 attempts to guess it. If the player can not find the correct number in 10 attempts, print "You lost. Want to play again? ".
- If the player types one of 'Y', 'YES', 'y', 'yes', 'ok' then start the game from the beginning.


> Hint: Use Python’s `random.randint()` to generate the number.
"""
import random

number=random.randint(1, 100)
count=0
while count<10:
    guess=int(input("Enter the guess between 1 and 100: "))
    count+=1
    if guess==number:
        print("You found the number")
        break
    elif guess>number:
        print("Too high")
    elif guess<number:
        print("Too low")
    else:
        print("Incorrect guess")
print(f"You found the number in {count} guesses.")
```

```python
"""
**5. Password Checker**
Task: Create a simple password checker.
- Ask the user to enter a password. 
- If the password is shorter than 8 characters, print "Password is too short." 
- If the password doesn't contain at least one uppercase letter, print "Password must contain an uppercase letter.". 
- If the password meets both criteria, print "Password is strong."

"""

def password_checker(password):
    if len(password)<8:
       print("Password is too short.")
    elif not any(c.isupper() for c in password):
       print("Password must contain an uppercase letter.")
    else:
       print("Password is strong")

password=input("Enter your password: ")
password_checker(password)
    
```

```python
"""
**6. Prime Numbers**
Task: Write a Python program that prints all prime numbers between 1 and 100.

> A prime number is a number greater than 1 that is not divisible by any number other than 1 and itself. Use nested loops to check divisibility.

---
"""
def print_primes(limit):
    for num in range(2, limit + 1):

        is_prime = True  # reset for each number

        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break

        if is_prime:
            print(num)


num = int(input("Till which number of primes you wanna see: "))
print_primes(num)

```

```python
import math

universities = [
    ['California Institute of Technology', 2175, 37704],
    ['Harvard', 19627, 39849],
    ['Massachusetts Institute of Technology', 10566, 40732],
    ['Princeton', 7802, 37000],
    ['Rice', 5879, 35551],
    ['Stanford', 19535, 40569],
    ['Yale', 11701, 40500]
]

# 1 Extract enrollment and tuition lists
def enrollment_stats(data):
    students = [uni[1] for uni in data]
    tuition = [uni[2] for uni in data]
    return students, tuition


# 2 Mean function
def mean(values):
    return sum(values) / len(values)


# 3 Median function
def median(values):
    sorted_values = sorted(values)
    n = len(sorted_values)
    mid = n // 2
    
    if n % 2 == 0:
        return (sorted_values[mid - 1] + sorted_values[mid]) / 2
    else:
        return sorted_values[mid]


# ---- MAIN LOGIC ----
students, tuition = enrollment_stats(universities)

total_students = sum(students)
total_tuition = sum(tuition)

student_mean = mean(students)
student_median = median(students)

tuition_mean = mean(tuition)
tuition_median = median(tuition)

print('*' * 30)
print(f"Total students: {total_students:,}")
print(f"Total tuition: $ {total_tuition:,}")
print()
print(f"Student mean: {student_mean:,.2f}")
print(f"Student median: {student_median:,}")
print()
print(f"Tuition mean: $ {tuition_mean:,.2f}")
print(f"Tuition median: $ {tuition_median:,}")
print('*' * 30)
```
